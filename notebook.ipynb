### Ages of Developers :

```python
ages = df['Age'].value_counts()
```

```python
plt.figure(figsize=(8,5))
ages.plot(kind='bar', color='blue')
plt.title("Ages of developers")
plt.xlabel("Ages")
plt.ylabel("Number of Developers")
plt.xticks(rotation=45)
plt.show()
```
